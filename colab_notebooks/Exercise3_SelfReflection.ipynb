# Exercise 3 — Self-Reflection Prompting

## Tools Used
- Google Colab
- ChatGPT

## Scenario
Demonstrating how an AI can improve its own response through structured self-reflection.

## Structure
Initial Answer → Reflection → Improved Answer

```python
# Prompt
prompt = "Explain why clear communication is important in a team."

print("Prompt:", prompt)
```

```python
# Initial AI response (simulated)

initial_response = """
Clear communication is important in a team because it helps people understand tasks and reduces confusion.
It ensures everyone knows their responsibilities and improves collaboration.
"""

print("Initial Response:")
print(initial_response)
```

```python
# Self-Reflection Step

reflection = """
Reflection on the initial response:

1. The explanation is too general.
2. It lacks real-world examples.
3. It does not mention consequences of poor communication.
4. It could include specific benefits like productivity, morale, and trust.

The response should be expanded with clearer reasoning and examples.
"""

print("Self-Reflection:")
print(reflection)
```

```python
# Improved Response After Reflection

improved_response = """
Clear communication is essential in a team because it ensures alignment, efficiency, and trust.

When team members clearly understand goals, deadlines, and responsibilities,
they can work more productively and avoid costly mistakes. For example,
miscommunication about deadlines can delay projects, while unclear expectations
can reduce morale and accountability.

Strong communication also builds trust. When team members openly share ideas
and feedback, collaboration improves and conflicts are resolved more quickly.

Without clear communication, teams may experience confusion, duplicated work,
low morale, and reduced performance.
"""

print("Improved Response:")
print(improved_response)
```
